```python
import momepy
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
buildings = gpd.read_file(momepy.datasets.get_path('bubenec'),
                          layer='buildings')
```

```python
f, ax = plt.subplots(figsize=(10, 10))
buildings.plot(ax=ax)
ax.set_axis_off()
plt.show()
```
